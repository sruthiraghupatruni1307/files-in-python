```python
# number of lines in a file
## number of words in a file
## number of chr in a file
with open('data.txt','r') as fg:
    k=fg.read()
    cnt=1
    for i in k:
        if i=="\n":
            cnt=cnt+1
    print(cnt)
```

```python
with open('data.txt','r') as fg:
    k=fg.read()
    cnt=1
    for i in k:
        if i=="\n":
            cnt=cnt+1
            wc=k.split()
    print(cnt)
    print(wc)
```
